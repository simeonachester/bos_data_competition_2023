```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
```

```python
path = "./MICS6_2023_Stats-Day-Competition/MICS6_2019_Stats Day Competition/Datasets/hh_Household_Stat_Day.xlsx"
dataset = pd.read_excel(path)
```

```python
# Mean number of HH members(HH48) by Region(HH7) then Area (HH6)
selected_columns = ["HH48", "HH7", "HH6"]
temp = dataset[selected_columns]
df_0 = temp[temp["HH48"].notna() & temp["HH48"].notnull()]
# df_0.info()

df_0[["HH48", "HH7"]].groupby(by="HH7").mean().round(2).sort_values(by="HH48")
order = df_0[["HH48", "HH7"]].groupby(
    by="HH7").mean().round(2).sort_values(by="HH48").index
```

```python

# fig, ax = plt.subplots(figsize=(15, 10))

ax = sns.catplot(
    data=df_0,
    x="HH7",
    y="HH48",
    hue="HH6",
    kind="boxen",
    height=10,
    order=order
)

# ax.set_xlabel("Region")
# ax.set_ylabel("Number of Household Members")
ax.set_xlabels(["Region"])
plt.title(
    f"Graph Showing Number of Household Members Grouped by Region and Area for {df_0['HH48'].count()} Households ordered by Regional Mean")
```
