```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
path = "./MICS6_2023_Stats-Day-Competition/MICS6_2019_Stats Day Competition/Datasets/hh_Household_Stat_Day.xlsx"
dataset = pd.read_excel(path)
```

```python
# Mean number of children(HH51 + HH52) per Household by Region(HH7) then Area (HH6)
selected_columns = ["HH51", "HH52", "HH7", "HH6"]
temp = dataset[selected_columns]
# temp.info()
df_0 = temp[temp["HH51"].notna() & temp["HH52"].notnull()]
df_0["total_children"] = df_0["HH51"] + df_0["HH52"]
# df_0.info()

# df_0[["HH7", "HH6", "total_children"]].groupby(
#    by=["HH7", "HH6"]).mean().round(2)  # .sort_values(by="total_children")
# order = df_0[["HH48", "HH7"]].groupby(
#    by="HH7").mean().round(2).sort_values(by="HH48").index
```

```python
sns.catplot(
    data=df_0,
    height=20,
    x="HH7",
    y="total_children",
    hue="HH6",
    kind="violin",
    # kind="swarm",
    split=True,
)
```
